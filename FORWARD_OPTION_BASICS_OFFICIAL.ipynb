```python
#Packages needed

#Numerical package
import numpy as np

#Graphical Package
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

We define our first payoffs.

```python
#We define a Long Call Option Payoff and Strike=200
def LongCall(S):
    return np.maximum(S-200,0)


#We define a Long Put Option Payoff and Strike=200
def LongPut(S):
    return np.maximum(200-S,0)
```

We make a plot with all the possible positions of a Call and Put option

```python
#We create a equispaced vector with
#values from 0 to 400
S=np.linspace(0,400)

#We create a figure to insert the plots
plt.figure(figsize =(15 ,10))

#In the first plot
plt.subplot(2,2,1)

#We write the title
plt.title("Long Call", fontsize=22)

#We write a x-axis label
plt.xlabel("Asset price at maturity", fontsize=18)

#We write a y-axis label
plt.ylabel("Payoff", fontsize=18)

#We do the plot of the Long Forward
plt.plot(S,LongCall(S), color="black")

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

#The second plot
plt.subplot(2,2,2)

#We write the title
plt.title("Short Call", fontsize=22)

#We write a x-axis label
plt.xlabel("Asset price at maturity", fontsize=18)

#We write a y-axis label
plt.ylabel("Payoff", fontsize=18)

#We do the plot of the Short Forward
plt.plot(S,-LongCall(S), color="black")

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

#In the first plot
plt.subplot(2,2,3)

#We write the title
plt.title("Long Put", fontsize=22)

#We write a x-axis label
plt.xlabel("Asset price at maturity", fontsize=18)

#We write a y-axis label
plt.ylabel("Payoff", fontsize=18)

#We do the plot of the Long Forward
plt.plot(S,LongPut(S), color="black")

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

#The second plot
plt.subplot(2,2,4)

#We write the title
plt.title("Short Put", fontsize=22)

#We write a x-axis label
plt.xlabel("Asset price at maturity", fontsize=18)

#We write a y-axis label
plt.ylabel("Payoff", fontsize=18)

#We do the plot of the Short Forward
plt.plot(S,-LongPut(S), color="black")

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.tight_layout()

plt.show()
```

But, sometimes we need to have more flexibility on our functions. Note that the Strike is fixed, we can define the call and put option, but this time the Strike is an input.

```python
#We define a Generic Long Call Option Payoff
def LongCallv2(S,K):
    return np.maximum(S-K,0)
```

```python
#We define a Generic Long Put Option Payoff
def LongPutv2(S,K):
    return np.maximum(K-S,0)
```

We use this new functions to draw a Bull Spread Strategy

```python
#We create a equispaced vector with
#values from 0 to 400
S=np.linspace(0,400)

#We create a figure to insert the plots
plt.figure(figsize =(15 ,10) )

#In the first plot
plt.subplot(2,2,1)

#We write the title
plt.title("Long Call Option with K=100", fontsize=22)

#We write a x-axis label
plt.xlabel("Asset price at maturity", fontsize=18)

#We write a y-axis label
plt.ylabel("Payoff", fontsize=18)

#We make the Long Call plot with Strike 200
plt.plot(S,LongCallv2(S,200), color="black")

#In the second plot
plt.subplot(2,2,2)

#We write the title
plt.title("Short Call Option with Strike K=250", fontsize=22)

#We write a x-axis label
plt.xlabel("Asset price at maturity", fontsize=18)

#We write a y-axis label
plt.ylabel("Payoff", fontsize=18)

#We make the Short Call plot  with Strike 250
plt.plot(S,-LongCallv2(S,250), color="black")

#In the third plot
plt.subplot(2,2,3)

#We write the title
plt.title("Long Call Option with K=200 + Short Call Option with Strike K=250", fontsize=22)

#We write a x-axis label
plt.xlabel("Asset price at maturity", fontsize=18)

#We write a y-axis label
plt.ylabel("Payoff", fontsize=18)

#We make the Strategy plot
plt.plot(S,LongCallv2(S,200)-LongCallv2(S,250), color="black")

#We make enough space between the subplots
plt.tight_layout()

plt.show()
```

STRATEGY: MARRIED PUT, PROTECTICE PUT

The married put and protective put strategies are identical, except for the time when the stock is acquired. The protective put involves buying a put to hedge a stock already in the portfolio. If the put is bought at the same time as the stock, the strategy is called a married put.
image.png

A Married Put is an options strategy where an investor holds a long position in a stock and buys a put option for protection. This strategy provides downside protection while allowing for potential upside gains. The payoff of this strategy combines the payoff of the stock position and the protective put option.

Here's a Python script to demonstrate the Married Put strategy using data visualization with Matplotlib:

1. Long Stock: Hold a position in the stock.
2. Buy Put Option: Buy a put option to protect against a decline in the stock price.
3. Plot the Payoff: Plot the payoff diagram for the long stock, the put option, and the combined Married Put strategy.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
stock_price = 100  # Current stock price
strike_price = 95  # Strike price of the put option
premium = 5        # Premium paid for the put option
shares = 1         # Number of shares/CONTRACT

# Define the range of possible stock prices at expiration
stock_prices = np.linspace(50, 150, 100)

# Calculate the payoff for the stock
stock_payoff = shares * (stock_prices - stock_price)

# Calculate the payoff for the put option
put_payoff = np.maximum(strike_price - stock_prices, 0) - premium

# Calculate the payoff for the Married Put (stock + put option)
married_put_payoff = stock_payoff + put_payoff

# Plot the payoffs
plt.figure(figsize=(10, 6))
plt.plot(stock_prices, stock_payoff, label='Stock Payoff', linestyle='--')
plt.plot(stock_prices, put_payoff, label='Put Option Payoff', linestyle='--')
plt.plot(stock_prices, married_put_payoff, label='Married Put Payoff', color='black')

# Highlight the break-even point
break_even = stock_price + premium
plt.axvline(break_even, color='red', linestyle='--', label='Break-even Price')

# Add titles and labels
plt.title('Married Put Payoff Diagram')
plt.xlabel('Stock Price at Expiration')
plt.ylabel('Profit / Loss')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

STRATEGY BEAR PUT SPREAD

The bear put spread strategy is another form of vertical spread. In this strategy, the investor simultaneously purchases put options at a specific strike price and also sells the same number of puts at a lower strike price. Both options are purchased for the same underlying asset and have the same expiration date.
2
 This strategy is used when the trader has a bearish sentiment about the underlying asset and expects the asset's price to decline. The strategy offers both limited losses and limited gains.

image.png

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
stock_price = 100  # Current stock price
strike_price_long_put = 120  # Strike price of the long put option
premium_long_put = 8         # Premium paid for the long put option
strike_price_short_put = 80  # Strike price of the short put option
premium_short_put = 3        # Premium received for the short put option

# Define the range of possible stock prices at expiration
stock_prices = np.linspace(50, 150, 100)

# Calculate the payoff for the long put option
long_put_payoff = np.maximum(strike_price_long_put - stock_prices, 0) - premium_long_put

# Calculate the payoff for the short put option
short_put_payoff = premium_short_put - np.maximum(strike_price_short_put - stock_prices, 0)

# Calculate the payoff for the Bear Put Spread (long put + short put)
bear_put_spread_payoff = long_put_payoff + short_put_payoff

# Plot the payoffs
plt.figure(figsize=(10, 6))
plt.plot(stock_prices, long_put_payoff, label='Long Put Payoff', linestyle='--')
plt.plot(stock_prices, short_put_payoff, label='Short Put Payoff', linestyle='--')
plt.plot(stock_prices, bear_put_spread_payoff, label='Bear Put Spread Payoff', color='black')

# Add titles and labels
plt.title('Bear Put Spread Payoff Diagram')
plt.xlabel('Stock Price at Expiration')
plt.ylabel('Profit / Loss')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

STRATEGY: PROTECTIVE COLLAR

A protective collar strategy is performed by purchasing an out-of-the-money (OTM) put option and simultaneously writing an OTM call option (of the same expiration) when you already own the underlying asset. This strategy is often used by investors after a long position in a stock has experienced substantial gains. This allows investors to have downside protection as the long put helps lock in the potential sale price. However, the trade-off is that they may be obligated to sell shares at a higher price, thereby forgoing the possibility of further profits.

An example of this strategy is if an investor is long on 100 shares of IBM at \$100 as of January 1. The investor could construct a protective collar by selling one IBM March 110 call and simultaneously buying one IBM March 95 put. The trader is protected below \$90 until the expiration date. The trade-off is that they may potentially be obligated to sell their shares at \$105 if IBM trades at that rate prior to expiry.

image.png

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
stock_price = 100  # Current stock price
strike_price_put = 90  # Strike price of the put option
premium_put = 5       # Premium paid for the put option
strike_price_call = 110  # Strike price of the call option
premium_call = 4       # Premium received for the call option

# Define the range of possible stock prices at expiration
stock_prices = np.linspace(50, 150, 100)

# Calculate the payoff for the long stock
long_stock_payoff = stock_prices - stock_price

# Calculate the payoff for the long put option
long_put_payoff = np.maximum(strike_price_put - stock_prices, 0) - premium_put

# Calculate the payoff for the short call option
short_call_payoff = premium_call - np.maximum(stock_prices - strike_price_call, 0)

# Calculate the payoff for the Protective Collar (long stock + long put - short call)
protective_collar_payoff = long_stock_payoff + long_put_payoff + short_call_payoff

# Plot the payoffs
plt.figure(figsize=(10, 6))
plt.plot(stock_prices, long_stock_payoff, label='Long Stock Payoff', linestyle='--')
plt.plot(stock_prices, long_put_payoff, label='Long Put Payoff', linestyle='--')
plt.plot(stock_prices, short_call_payoff, label='Short Call Payoff', linestyle='--')
plt.plot(stock_prices, protective_collar_payoff, label='Protective Collar Payoff', color='black')

# Add vertical lines for strike prices
plt.axvline(strike_price_put, color='red', linestyle='--', label='Put Strike Price')
plt.axvline(strike_price_call, color='blue', linestyle='--', label='Call Strike Price')

# Add titles and labels
plt.title('Protective Collar Payoff Diagram')
plt.xlabel('Stock Price at Expiration')
plt.ylabel('Profit / Loss')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

STRATEGY: Long Straddle

A long straddle options strategy occurs when an investor simultaneously purchases a call and put option on the same underlying asset with the same strike price and expiration date. An investor will often use this strategy when they believe the price of the underlying asset will move significantly out of a specific range, but they are unsure of which direction the move will take.

Theoretically, this strategy allows the investor to have the opportunity for unlimited gains. At the same time, the maximum loss this investor can experience is limited to the cost of both options contracts combined.

image.png

A Long Straddle is an options strategy where an investor buys both a call option and a put option with the same strike price and expiration date. This strategy profits from significant movement in the stock price, either up or down, while the maximum loss is limited to the total premium paid for both options.

Here's a Python script to demonstrate the Long Straddle strategy using data visualization with Matplotlib:

Buy Call Option: Purchase a call option with a specific strike price.
Buy Put Option: Purchase a put option with the same strike price.
Plot the Payoff: Plot the payoff diagram for the call option, put option, and the combined Long Straddle strategy.

K biết lên cửa nào

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
stock_price = 100  # Current stock price
strike_price = 100  # Strike price for both call and put options
premium_call = 6   # Premium paid for the call option
premium_put = 5    # Premium paid for the put option

# Define the range of possible stock prices at expiration
stock_prices = np.linspace(50, 150, 100)

# Calculate the payoff for the call option
call_payoff = np.maximum(stock_prices - strike_price, 0) - premium_call

# Calculate the payoff for the put option
put_payoff = np.maximum(strike_price - stock_prices, 0) - premium_put

# Calculate the payoff for the Long Straddle (call + put)
long_straddle_payoff = call_payoff + put_payoff

# Plot the payoffs
plt.figure(figsize=(10, 6))
plt.plot(stock_prices, call_payoff, label='Call Option Payoff', linestyle='--')
plt.plot(stock_prices, put_payoff, label='Put Option Payoff', linestyle='--')
plt.plot(stock_prices, long_straddle_payoff, label='Long Straddle Payoff', color='black')

# Add vertical line for the strike price
plt.axvline(strike_price, color='red', linestyle='--', label='Strike Price')

# Add titles and labels
plt.title('Long Straddle Payoff Diagram')
plt.xlabel('Stock Price at Expiration')
plt.ylabel('Profit / Loss')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

A Long Strangle is an options strategy where an investor buys both a call option and a put option with different strike prices but with the same expiration date. This strategy profits from significant movement in the stock price, either up or down, while the maximum loss is limited to the total premium paid for both options.

Here's a Python script to demonstrate the Long Strangle strategy using data visualization with Matplotlib:

Buy Call Option: Purchase a call option with a higher strike price.
Buy Put Option: Purchase a put option with a lower strike price.
Plot the Payoff: Plot the payoff diagram for the call option, put option, and the combined Long Strangle strategy.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
stock_price = 100        # Current stock price (for context)
strike_price_call = 110  # Strike price of the call option
premium_call = 5         # Premium paid for the call option
strike_price_put = 90    # Strike price of the put option
premium_put = 4          # Premium paid for the put option

# Define the range of possible stock prices at expiration
stock_prices = np.linspace(50, 150, 100)

# Calculate the payoff for the call option
call_payoff = np.maximum(stock_prices - strike_price_call, 0) - premium_call

# Calculate the payoff for the put option
put_payoff = np.maximum(strike_price_put - stock_prices, 0) - premium_put

# Calculate the payoff for the Long Strangle (call + put)
long_strangle_payoff = call_payoff + put_payoff

# Plot the payoffs
plt.figure(figsize=(10, 6))
plt.plot(stock_prices, call_payoff, label='Call Option Payoff', linestyle='--')
plt.plot(stock_prices, put_payoff, label='Put Option Payoff', linestyle='--')
plt.plot(stock_prices, long_strangle_payoff, label='Long Strangle Payoff', color='black')

# Add vertical lines for strike prices
plt.axvline(strike_price_call, color='red', linestyle='--', label='Call Strike Price')
plt.axvline(strike_price_put, color='blue', linestyle='--', label='Put Strike Price')

# Add titles and labels
plt.title('Long Strangle Payoff Diagram')
plt.xlabel('Stock Price at Expiration')
plt.ylabel('Profit / Loss')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
